```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r'M:/Risk Management/DW/TUX Scorecard/Prod Testing/NullReplacementLogicTesting/BeforeNullReplacementLogicUpdated_11012024-07292025-scoring.csv')
```

```python
df2 = pd.read_csv(r'M:/Risk Management/DW/TUX Scorecard/Prod Testing/NullReplacementLogicTesting/AfterNullReplacementLogicUpdated_11012024-07292025-scoring.csv')
```

```python
df["RowId"] = df["RowId"].astype(str).str.replace("'", "", regex=False)
df2["RowId"] = df2["RowId"].astype(str).str.replace("'", "", regex=False)
```

```python
df.info()
```

```python
df2.info()
```

```python
cols_to_keep = [
    'AccountId', 'ApplicationScore', 'ApplicationProb2Def', 'ApplicationIsApproved', 'ApplicationCreditProfileType',
    'ApplicationTier'
]
```

```python
df = df[cols_to_keep].copy()
df2 = df2[cols_to_keep].copy()
```

```python
df = df.drop_duplicates(subset='AccountId')
df2 = df2.drop_duplicates(subset='AccountId')
```

```python
score_cols = ['ApplicationScore', 'ApplicationProb2Def']
```

```python
non_num_cols = ['ApplicationCreditProfileType', 'ApplicationTier', 'ApplicationIsApproved']
```

```python
merged_df = df.merge(df2, on='AccountId', suffixes=('_before', '_after'))
```

```python
for col in score_cols:
    merged_df[f'{col}_diff'] = merged_df[f'{col}_before'] - merged_df[f'{col}_after']
```

```python
for col in non_num_cols:
    col_before = f'{col}_before'
    col_after = f'{col}_after'
    merged_df[f'{col}_changed'] = (merged_df[col_before] != merged_df[col_after]).astype(int)
```

```python
output_path = r'M:/Risk Management/DW/TUX Scorecard/Prod Testing/NullReplacementLogicTesting/ComparisonFileSimple.xlsx'
merged_df.to_excel(output_path, index=False)
```
