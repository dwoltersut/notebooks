```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'M:/Risk Management/DW/Scorecard/TUX_Scores/Oct2024_FormattedSampleData_Scoring_DaveS.csv')
```

```python
df2 = pd.read_csv(r'M:/Risk Management/DW/Scorecard/TUX_Scores/PARSED_PAYLOADS_DBEAVERData_OpenBKReparse.csv')
```

```python
df1['BIGDEBTORID'] = df1['id'].astype(str).str[7:15]
```

```python
df1['ACCOUNTID'] = pd.to_numeric(df1['ACCOUNTID'], errors='coerce').astype('Int64')
df1['BIGDEBTORID'] = pd.to_numeric(df1['BIGDEBTORID'], errors='coerce').astype('Int64')
```

```python
df2['ACCOUNTID'] = df2['ACCOUNTID'].str.replace('"', '', regex=False)
```

```python
df2['DEBTORSTATE'] = df2['DEBTORSTATE'].str.replace('"', '', regex=False)
```

```python
df2['BITAPPROVED__APP'] = df2['BITAPPROVED__APP'].str.replace('"', '', regex=False)
```

```python
df2['BITFUNDED__APP'] = df2['BITFUNDED__APP'].str.replace('"', '', regex=False)
```

```python
df2.info()
```

```python
df2['ACCOUNTID'] = pd.to_numeric(df2['ACCOUNTID'], errors='coerce').astype('Int64')
df2['BIGDEBTORID'] = pd.to_numeric(df2['BIGDEBTORID'], errors='coerce').astype('Int64')
```

```python
df2.head()
```

```python
df2.drop(columns=['TUXSC1_OPENBK', 'TUXSC2_THIN', 'TUXSC3_THICKDIRTY', 'TUXSC4_THICKCLEAN', 'NEWCREDITSCORECARDNAME'], inplace=True)
```

```python
df3.info()
```

```python
df3.rename(columns={'bigaccountid':'ACCOUNTID', 'bigdebtorid':'BIGDEBTORID', 'intType':'OpenBKType'}, inplace=True)
```

```python
df3['OpenBKType'] = pd.to_numeric(df3['OpenBKType'], errors='coerce').astype(int)
```

```python
df3.head()
```

```python
df_newbkapp_sc = pd.merge(df2, df3, on=['ACCOUNTID', 'BIGDEBTORID'], how='left')
```

```python
df_newbkapp_sc.info()
```

```python
df_newbkapp_sc['OpenBKType'] = df_newbkapp_sc['OpenBKType'].fillna(0).astype(int)
```

```python
df_newbkapp_sc.head()
```

```python
segment = [
    (df_newbkapp_sc['OpenBKType'].isin([7, 13])),
    (df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] < 2) | (df_newbkapp_sc['RE01S'].isna())),
    (df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] >= 2) & (df_newbkapp_sc['G251B'] > 1)),
    (df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] >= 2) & (df_newbkapp_sc['G251B'] <= 1))
]
```

```python
values = ['Scorecard1', 'Scorecard2', 'Scorecard3', 'Scorecard4']
```

```python
df_newbkapp_sc['NewCreditScorecardName'] = np.select(segment, values, default='Unknown')
```

```python
df_newbkapp_sc['TUXSC1_OpenBK'] = np.where(df_newbkapp_sc['OpenBKType'].isin([7, 13]), 1, 0)
```

```python
df_newbkapp_sc['TUXSC2_Thin'] = np.where((df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] < 2) | (df_newbkapp_sc['RE01S'].isna())), 1, 0)
```

```python
df_newbkapp_sc['TUXSC3_ThickDirty'] = np.where((df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] >= 2) & (df_newbkapp_sc['G251B'] > 1)), 1, 0)
```

```python
df_newbkapp_sc['TUXSC4_ThickClean'] = np.where((df_newbkapp_sc['OpenBKType'] == 0) & ((df_newbkapp_sc['RE01S'] >= 2) & (df_newbkapp_sc['G251B'] <= 1)), 1, 0)
```
