```python
def plot_avg_score_by_feat(df, feature_col, score_col, bin_size):
    
    # Create new column for bins from Total_Score_Scaled 
    df['score_bin'] = (df[score_col] // bin_size) * bin_size  
    
    # Calculate the average `Feature Value` for each score bin
    average_scores = df.groupby('score_bin')[feature_col].mean().reset_index()
    
    # Plot 
    plt.figure(figsize=(15, 8))
    plt.plot(
        average_scores['score_bin'], 
        average_scores[feature_col],  
        marker='o', linestyle='-', color='b')

    plt.xlabel(f'Score Range (Grouped by {bin_size})')  
    plt.ylabel(f'Average {feature_col}')  
    plt.title(f'{feature_col} by Score Group')
    plt.grid(axis='y', linestyle='--', alpha=0.7)  

    for spine in plt.gca().spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(2)

    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_feature_vs_score(df, feature_list, score_col="Total_Score_Scaled", num_cols=3, figsize=(15, 10), alpha=0.5):
    """
    Parameters:
    - feature_list: Original model features (before WoE binning)
    - score_col: Final transformed score col (Total_Score_Scaled)
    - num_cols: Number of cols in the subplot grid 
    - alpha: Transparency for scatter points (default: 0.5)
    """

    num_features = len(feature_list)
    num_rows = (num_features // num_cols) + (num_features % num_cols > 0)  # Calculate rows dynamically

    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=figsize)
    axes = axes.flatten()  # Flatten grid to iterate easily

    for i, feature in enumerate(feature_list):
        ax = axes[i]
        sns.regplot(data=df, x=feature, y=score_col, scatter_kws={'alpha': alpha}, line_kws={"color": "red"}, ax=ax)
        ax.set_title(f"{feature} vs. {score_col}")
        ax.set_xlabel(feature)
        ax.set_ylabel(score_col)

    # Hide any empty subplots (if num_features < num_rows*num_cols)
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
list_vars_reduced = list_vars_reduced

# df1 = Train
# df2 = Valid
# df3 = Test

# Plot for Training Data
plot_feature_vs_score(df1, list_vars_reduced)
```

```python
# Plot for Validation Data
plot_feature_vs_score(df2, list_vars_reduced)
```

```python
# Plot for Test Data
plot_feature_vs_score(df3, list_vars_reduced)
```
